```python
import os
import pandas as pd

# Set the path to the folder containing .txt files
folder_path = 'aggTrade'  # ← Change this to your folder path

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        txt_path = os.path.join(folder_path, filename)
        csv_path = os.path.join(folder_path, filename.replace(".txt", ".csv"))
        
        try:
            # Read the .txt as CSV
            df = pd.read_csv(txt_path)
            # Save it as .csv
            df.to_csv(csv_path, index=False)
            print(f"Converted: {filename} → {os.path.basename(csv_path)}")
        except Exception as e:
            print(f"Failed to convert {filename}: {e}")

# Merge all .csv files
merged_df = pd.DataFrame()
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        csv_path = os.path.join(folder_path, filename)
        try:
            df = pd.read_csv(csv_path)
            merged_df = pd.concat([merged_df, df], ignore_index=True)
        except Exception as e:
            print(f"Failed to read {filename}: {e}")

# Save the merged CSV
merged_path = os.path.join(folder_path, "aggTrade-tot.csv")
merged_df.to_csv(merged_path, index=False)
print(f"\n✅ All CSVs merged into: {merged_path}")
```
